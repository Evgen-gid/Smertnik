```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Загрузка данных
df = pd.read_csv('HR.csv')

# 1. Основные статистики для количественных переменных
numeric_cols = ['satisfaction_level', 'last_evaluation', 'number_project',
                'average_montly_hours', 'time_spend_company']
stats_df = df[numeric_cols].describe().transpose()
stats_df['mode'] = df[numeric_cols].mode().transpose()[0]
stats_df = stats_df[['mean', '50%', 'mode', 'min', 'max', 'std']]
stats_df.columns = ['mean', 'median', 'mode', 'min', 'max', 'std_dev']
print("Основные статистики:\n", stats_df)

# 2. Корреляционная матрица
corr_matrix = df[numeric_cols].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()

# Нахождение самых скоррелированных и наименее скоррелированных пар
corr_pairs = corr_matrix.unstack().sort_values(ascending=False)
unique_pairs = corr_pairs[corr_pairs.index.get_level_values(0) != corr_pairs.index.get_level_values(1)]
print("\nСамые скоррелированные переменные:", unique_pairs.index[0])
print("Наименее скоррелированные переменные:", unique_pairs.index[-1])

# 3. Количество сотрудников по департаментам
dept_counts = df['department'].value_counts()
print("\nКоличество сотрудников по департаментам:\n", dept_counts)

# 4. Распределение по зарплатам
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='salary')
plt.title('Распределение сотрудников по зарплатам')
plt.show()

# Распределение по зарплатам в разрезе департаментов
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='department', hue='salary')
plt.xticks(rotation=45)
plt.title('Распределение зарплат по департаментам')
plt.show()

# 5. Проверка гипотезы о времени работы и зарплате
high_salary = df[df['salary'] == 'high']['average_montly_hours']
low_salary = df[df['salary'].isin(['low', 'medium'])]['average_montly_hours']
t_stat, p_value = stats.ttest_ind(high_salary, low_salary)
print(f"\nРезультат t-теста: p-value = {p_value:.4f}")
print("Гипотеза", "подтверждается" if p_value < 0.05 else "не подтверждается")

# 6. Показатели для уволившихся и не уволившихся
left_0 = df[df['left'] == 0]
left_1 = df[df['left'] == 1]

metrics = {
    'left=0': [
        left_0['promotion_last_5years'].mean(),
        left_0['satisfaction_level'].mean(),
        left_0['number_project'].mean()
    ],
    'left=1': [
        left_1['promotion_last_5years'].mean(),
        left_1['satisfaction_level'].mean(),
        left_1['number_project'].mean()
    ]
}

metrics_df = pd.DataFrame(metrics, index=['Доля повышений', 'Средняя удовлетворенность', 'Среднее проектов'])
print("\nПоказатели по группам:\n", metrics_df)
```
